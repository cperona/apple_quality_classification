## Apple quality classification
Import the dataset from: https://www.kaggle.com/datasets/nelgiriyewithana/apple-quality

```python
import numpy as np
```

```python
file_path = "apple_quality_dataset.csv"

data = np.genfromtxt(file_path, delimiter=',')
data = data[1:-1,1:-1]

X = data
print(X)

tags = np.genfromtxt(file_path, delimiter=',', dtype=str)
tags = tags[1:-1,-1]
print(tags)

def goodbad(label):
    if label == "good":
        return np.int32(1)
    else:
        return np.int32(0)

y = [goodbad(label) for label in tags]
print(y)
```

```python
import pandas as pd

X_df = pd.DataFrame(data=X, columns=["Size","Weight","Sweetness","Crunchiness","Juiciness","Ripeness","Acidity"])
```

```python
X_df
```

Split train and test

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

Use a decision tree classifier to identify the quality

```python
from sklearn import tree
from sklearn.metrics import accuracy_score

classifier = tree.DecisionTreeClassifier()
classifier.fit(X_train, y_train)
y_pred = classifier.predict(X_test)

acc = accuracy_score(y_test, y_pred)
print(acc)
```

Train the model with the best five features to improve efficiency

```python
from sklearn.feature_selection import SelectKBest

#Select k amount of features.
feature_selector = SelectKBest(k=5)

X_train_selected = feature_selector.fit_transform(X_train, y_train)
X_test_selected = feature_selector.transform(X_test)

#Create a Decision Tree Classifier
classifier2 = tree.DecisionTreeClassifier()

classifier2.fit(X_train_selected, y_train)
y_pred = classifier2.predict(X_test_selected)

print(y_pred)
```

Show selected feature names

```python
import numpy as np

selected_feature_mask = feature_selector.get_support()
selected_feature_names = np.array(["Size","Weight","Sweetness","Crunchiness","Juiciness","Ripeness","Acidity"])[selected_feature_mask]

print(selected_feature_names)
```
